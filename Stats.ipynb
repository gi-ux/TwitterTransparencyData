```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import plotly.graph_objects as go
import datetime
import plotly.express as px
import time
from time import perf_counter
from statistics import mean
import glob
import sys
sys.path.append('C:/Users/gianl/Desktop/Gi/Supsi/BachelorProject/notebook/tweet_utils')
import tweets_utils
import importlib
import seaborn as sns
from tqdm.notebook import tqdm
folder = "F:/Supsi/TwitterData/"
```

```python
repos = []
n_users = []
n_tweets = []
mean = []
```

```python
repo = glob.glob(folder + "*")
cont = 0
for i in repo:
#     if cont == 5:
#         break
#     cont = cont + 1
    repos.append(i)
    number_of_tweets = 0
    names = []
    print(i)
    path_profiles = glob.glob(i + "/*users_csv*.csv")
    path_data = glob.glob(i + "/*tweets_csv*.csv")
    users = pd.read_csv(path_profiles[0], lineterminator="\n", low_memory=False)
    tweets = pd.DataFrame()
    for file in path_data:
        for chunk in pd.read_csv(file, lineterminator="\n", low_memory=False, chunksize=int(1e6)):
            names.extend(list(chunk["user_display_name"]))
            number_of_tweets = number_of_tweets + len(chunk)
    median = round(pd.Series(names).value_counts().median(),0)
    print(f"Number of users: {len(users)}")
    n_users.append(len(users))
    print(f"Number of tweets: {number_of_tweets}")
    n_tweets.append(number_of_tweets)
    print(f"Number of tweets per users: {median}")
    mean.append(median)
```

```python
df = pd.DataFrame(list(zip(repos, n_users, n_tweets, mean)), 
                       columns = ["Name", "Users number", "Tweets number", "Median"])
df
```
